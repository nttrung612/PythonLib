```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,8])
y = np.array([3,10])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

Default X-Points

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
# xpoints here = [0, 1, 2, 3, 4, 5] as default
plt.plot(ypoints)
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

Format Strings fmt

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-r') #marker-line-color (search: line syntax, color syntax)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20) #ms: marker size
plt.show()
```

Marker Color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r') #mec: marker edge color
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r') #mfc: marker face color
plt.show()
```

Matplotlib Line

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = 'dotted') #linestyle: solid, dashed, dashdot, dotted
plt.show()
```

Line Color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

Multiple Lines

Matplotlib Labels and Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.show()
```

Matplotlib Adding Grid Lines

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid() #can add axis = 'x' or 'y' or 'both' and color, linestyle, linewidth as arguments

plt.show()
```

Matplotlib Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y) #can add color, size, alpha, marker, linewidths, edgecolors as arguments

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

Matplotlib Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1) #vertical bar chart
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height =0.1) #horizontal bar chart
plt.show()
```

Histograms

```python
x = np.random.normal(170, 10, 250) #generate 250 random numbers with a mean of 170, and standard deviation of 10

plt.hist(x) #histogram
plt.show()
```

Pie Charts

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels) #pie chart, can change start angle
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
#explode: pull out a slice, shadow: add shadow to the pie chart
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
